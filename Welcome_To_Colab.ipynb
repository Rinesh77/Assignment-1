```
# Problem 1:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Assignment1/WHR-2024-5CS037.csv")

# Display the first 10 rows
print(df.head(10))

# Identify the number of rows and columns
print(df.shape)

# List all columns and their data types
print(df.dtypes)
```

```

# Basic Statistics
# Calculate the mean, median, and standard deviation for the Score column
print(df['score'].mean(), df['score'].median(), df['score'].std())

# Identify the country with the highest and lowest happiness scores
print(df.loc[df['score'].idxmax()], df.loc[df['score'].idxmin()])
```

```
# Missing Values
# Check if there are any missing values
print(df.isnull().sum())
```

```
#  Filtering and Sorting
# Filter countries with a Score greater than 7.5
filtered_df = df[df['score'] > 7.5] # Changed 'Score' to 'score'

# Sort by GDP per Capita in descending order and display top 10 rows
sorted_df = filtered_df.sort_values(by='Log GDP per capita', ascending=False)
print(sorted_df.head(10))
```

```
# Adding New Columns
# Create a new column 'Happiness Category'
df['Happiness Category'] = pd.cut(df['score'], bins=[-np.inf, 4, 6, np.inf], labels=['Low', 'Medium', 'High'])
```

```
# Problem2:
# Data Visualizations
# Bar Plot: Top 10 happiest countries by Score
top_10_happiest = df.nlargest(10, 'score')
# Changed 'Score' to 'score' for the y-axis
top_10_happiest.plot(kind='bar', x='Country name', y='score', title='Top 10 Happiest Countries')
```

```
#Line Plot: Top 10 unhappiest countries by Score
top_10_unhappiest = df.nsmallest(10, 'score')
top_10_unhappiest.plot(kind='line', x='Country name', y='score', title='Top 10 Unhappiest Countries')
# Histogram for the Score column
df['score'].plot(kind='hist', bins=20, title='Score Distribution')

#Scatter Plot: GDP per Capita vs Score
# Check for the correct column name in your DataFrame using df.columns
print(df.columns)

# Replace 'Log GDP per Capita' with the actual column name from your DataFrame if it's different
df.plot(kind='scatter', x='Log GDP per capita', y='score', title='Log GDP per Capita vs score')
```

```
# Advanced Data Exploration - South Asia Dataset
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
south_asia = df[df['Country name'].isin(south_asia_countries)]
south_asia.to_csv("south_asia_data.csv", index=False)

# Check the actual column name in your DataFrame
print(south_asia.columns)

# Composite Score Ranking
# Replace 'Healthy life expectancy at birth' with the correct column name
# Assuming the correct column name is 'Healthy life expectancy'
south_asia['Composite Score'] = 0.40 * south_asia['Log GDP per capita'] + 0.30 * south_asia['Social support'] + 0.30 * south_asia['Healthy life expectancy']
south_asia = south_asia.sort_values(by='Composite Score', ascending=False)

# Top 5 countries based on Composite Score
south_asia.head(5).plot(kind='barh', x='Country name', y='Composite Score', title='Top 5 Countries by Composite Score')
```

```
# Outlier Detection using 1.5*IQR rule
Q1 = south_asia['score'].quantile(0.25)
Q3 = south_asia['score'].quantile(0.75)
IQR = Q3 - Q1
outliers = south_asia[(south_asia['score'] < (Q1 - 1.5 * IQR)) | (south_asia['score'] > (Q3 + 1.5 * IQR))]

# Scatter plot with GDP per Capita vs Score highlighting outliers
# Changed 'Log GDP per Capita' to 'Log GDP per capita' and 'Score' to 'score'
plt.scatter(south_asia['Log GDP per capita'], south_asia['score'], label='Normal')
# Changed 'GDP per Capita' to 'Log GDP per capita' and 'Score' to 'score' for outlier points
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers')
plt.xlabel('GDP per Capita')
plt.ylabel('score')
plt.title('Outliers in GDP per Capita vs Score')
plt.legend()
plt.show()
```

```
# Correlation Analysis: Freedom to Make Life Choices and Generosity vs Score
south_asia.plot(kind='scatter', x='Freedom to make life choices', y='score', title='Freedom to Make Life Choices vs Score') # Changed 'Score' to 'score'
south_asia.plot(kind='scatter', x='Generosity', y='score', title='Generosity vs Score')
```

```
# Adding GDP-Score Gap Column
# Replace 'GDP per Capita' with the actual column name from your DataFrame, likely 'Log GDP per capita'
south_asia['GDP-Score Gap'] = south_asia['Log GDP per capita'] - south_asia['score']

# Rank by GDP-Score Gap
ascending_gap = south_asia.sort_values(by='GDP-Score Gap')
descending_gap = south_asia.sort_values(by='GDP-Score Gap', ascending=False)

# Top 3 positive and negative gaps
top_positive_gap = descending_gap.head(3)
top_negative_gap = ascending_gap.head(3)

# Bar chart for top positive and negative gaps
plt.bar(top_positive_gap['Country name'], top_positive_gap['GDP-Score Gap'], color='green', label='Top 3 Positive Gaps')
plt.bar(top_negative_gap['Country name'], top_negative_gap['GDP-Score Gap'], color='red', label='Top 3 Negative Gaps')
plt.title('Top 3 Positive and Negative GDP-Score Gaps')
plt.xlabel('Country')
plt.ylabel('GDP-Score Gap')
plt.legend()
plt.tight_layout()
plt.show()
```

```
# Comparative Analysis - Middle Eastern Dataset
middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]
middle_east = df[df['Country name'].isin(middle_east_countries)]
middle_east.to_csv("middle_east_data.csv", index=False)
```

```
# Descriptive Statistics for South Asia and Middle East
south_asia_mean = south_asia['score'].mean()
middle_east_mean = middle_east['score'].mean()
print("South Asia Mean Score:", south_asia_mean, "Middle East Mean Score:", middle_east_mean)
```

```
# Top and Bottom Performers
south_asia_top_3 = south_asia.nlargest(3, 'score')
middle_east_top_3 = middle_east.nlargest(3, 'score')
south_asia_bottom_3 = south_asia.nsmallest(3, 'score')
middle_east_bottom_3 = middle_east.nsmallest(3, 'score')

# Bar charts comparing top and bottom performers
# Changed 'Score' to 'score' for the y-axis
plt.bar(south_asia_top_3['Country name'], south_asia_top_3['score'], label='South Asia Top 3', color='blue')
# Changed 'Score' to 'score' for the y-axis
plt.bar(middle_east_top_3['Country name'], middle_east_top_3['score'], label='Middle East Top 3', color='orange')
plt.legend()
plt.title('Top Performers in South Asia and Middle East')
plt.ylabel('Score')
plt.show()

# Changed 'Score' to 'score' for the y-axis
plt.bar(south_asia_bottom_3['Country name'], south_asia_bottom_3['score'], label='South Asia Bottom 3', color='red')
# Changed 'Score' to 'score' for the y-axis
plt.bar(middle_east_bottom_3['Country name'], middle_east_bottom_3['score'], label='Middle East Bottom 3', color='green')
plt.legend()
plt.title('Bottom Performers in South Asia and Middle East')
plt.ylabel('Score')
plt.show()
```

```
# Metric Comparisons - GDP per Capita, Social Support, Healthy Life Expectancy
south_asia_avg = south_asia[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()
middle_east_avg = middle_east[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()

print("South Asia Averages:", south_asia_avg)
print("Middle East Averages:", middle_east_avg)
```

```
# Grouped Bar Chart for Metric Comparisons
metrics = ['GDP per Capita', 'Social Support', 'Healthy Life Expectancy']
south_asia_avg.plot(kind='bar', label='South Asia')
middle_east_avg.plot(kind='bar', label='Middle East')
plt.title('Metric Comparison between South Asia and Middle East')
plt.ylabel('Average Value')
plt.xticks(ticks=np.arange(len(metrics)), labels=metrics, rotation=0)
plt.legend()
plt.show()
```

```
# Happiness Disparity - Range and Coefficient of Variation (CV)
south_asia_range = south_asia['score'].max() - south_asia['score'].min()
middle_east_range = middle_east['score'].max() - middle_east['score'].min()
south_asia_cv = south_asia['score'].std() / south_asia['score'].mean()
middle_east_cv = middle_east['score'].std() / middle_east['score'].mean()

print("South Asia Range:", south_asia_range, "Middle East Range:", middle_east_range)
print("South Asia CV:", south_asia_cv, "Middle East CV:", middle_east_cv)
```

```
# Correlation Analysis for Score with Freedom to Make Life Choices and Generosity
# Changed 'Score' to 'score' to match the actual column name in the DataFrame
south_asia_corr_life_choices = south_asia['score'].corr(south_asia['Freedom to make life choices'])
middle_east_corr_life_choices = middle_east['score'].corr(middle_east['Freedom to make life choices'])

south_asia_corr_generosity = south_asia['score'].corr(south_asia['Generosity'])
middle_east_corr_generosity = middle_east['score'].corr(middle_east['Generosity'])

print("South Asia Freedom to Make Life Choices Correlation:", south_asia_corr_life_choices)
print("Middle East Freedom to Make Life Choices Correlation:", middle_east_corr_life_choices)
print("South Asia Generosity Correlation:", south_asia_corr_generosity)
print("Middle East Generosity Correlation:", middle_east_corr_generosity)
```

```
# Scatter Plots for Correlations
south_asia.plot(kind='scatter', x='Freedom to make life choices', y='score', title='South Asia - Freedom to Make Life Choices vs Score')
middle_east.plot(kind='scatter', x='Freedom to make life choices', y='score', title='Middle East - Freedom to Make Life Choices vs Score')
south_asia.plot(kind='scatter', x='Generosity', y='score', title='South Asia - Generosity vs Score')
middle_east.plot(kind='scatter', x='Generosity', y='score', title='Middle East - Generosity vs Score')
```

```
# Outlier Detection based on Score and GDP per Capita
def find_outliers_south_asia(column_name):
    Q1 = south_asia[column_name].quantile(0.25)
    Q3 = south_asia[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return south_asia[(south_asia[column_name] < lower_bound) | (south_asia[column_name] > upper_bound)]
def find_outliers_middle_east(column_name):
    # Changed 'middle_east_data' to 'middle_east' to use the correct DataFrame
    Q1 = middle_east[column_name].quantile(0.25)
    Q3 = middle_east[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Changed 'middle_east_data' to 'middle_east' to use the correct DataFrame
    return middle_east[(middle_east[column_name]<lower_bound) | (middle_east[column_name]>upper_bound)]
if find_outliers_south_asia('score').empty:
  print("No Outliers")
else:
  print(find_outliers_south_asia('score'))

if find_outliers_middle_east('score').empty:
  print("No Outliers")
else:
  print(find_outliers_middle_east('score'))

if find_outliers_south_asia('Log GDP per capita').empty:
  print("No Outliers")
else:
  print(find_outliers_south_asia('Log GDP per capita'))

if find_outliers_middle_east('Log GDP per capita').empty:
  print("No Outliers")
else:
  print(find_outliers_middle_east('Log GDP per capita'))

#Plotting the outliers for each region
south_asia_normal_plot = south_asia.plot(kind='scatter', x='Log GDP per capita', y='score', title='GDP per Capita vs Score South Asia')

middle_east_normal_plot = middle_east.plot(kind='scatter', x='Log GDP per capita', y='score', title='GDP per Capita vs Score Middle East')
outliers_south_asia = pd.concat([find_outliers_south_asia('score'), find_outliers_south_asia('Log GDP per capita')])

outliers_middle_east = pd.concat([find_outliers_middle_east('score'), find_outliers_middle_east('Log GDP per capita')])
south_asia_normal_plot.scatter(outliers_south_asia['score'], outliers_south_asia['Log GDP per capita'], color='red', label='Outliers')
middle_east_normal_plot.scatter(outliers_middle_east['score'], outliers_middle_east['Log GDP per capita'], color='red', label='Outliers')
plt.legend()
```
